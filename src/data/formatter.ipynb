```python
import pandas as pd
import re

# Load the CSV file
file_path = 'F2021.csv'
df = pd.read_csv(file_path)

# Function to convert time format
def convert_time_format(time_str):
    # Convert times like '11:00a' to '11:00AM' and '12:00p' to '12:00PM'
    def convert(match):
        time, period = match.groups()
        period = period.upper() + 'M'
        return f"{time}{period}"
    
    return re.sub(r'(\d{1,2}:\d{2})([ap])', convert, time_str)

# Function to convert day format
def convert_days(day_str):
    # Convert 'Th' to 'R' and 'SSu' to 'SU'
    day_str = day_str.replace('Th', 'R')
    day_str = day_str.replace('SSu', 'SU')
    return day_str

# Apply the conversion function to the 'Day and Time' column
df['Day and Time'] = df['Day and Time'].apply(convert_time_format)
df['Day and Time'] = df['Day and Time'].apply(convert_days)

# Save the modified CSV file
output_file_path = 'F2021_reformatted.csv'
df.to_csv(output_file_path, index=False)

df.head()
```
